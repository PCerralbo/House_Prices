## Some ideas before start
This is a regression problem. There are several models to apply: multiple linear regression (linear regression, lasso, ridge, elastic net), SVR (suport vector regression), Random Forest, ....

The first one (linear regressions) need to accomplish different assumptions that we will need to check:
* Linearity ( Correct functional form )
* Homoscedasticity ( Constant Error Variance )( vs Heteroscedasticity ).
* Independence of Errors ( vs Autocorrelation )
* Multivariate Normality ( Normality of Errors )
* No or little Multicollinearity.

We will check these assumptions and prepare the data to linear models

```python
def plotting_3_chart(df, feature):
    ''' This function plots histogram, QQ_plot and Boxplot.
    https://www.kaggle.com/masumrumi/a-detailed-regression-guide-with-house-pricing'''
    ## Importing seaborn, matplotlab and scipy modules. 
    import matplotlib.gridspec as gridspec
    import matplotlib.style as style
    style.use('ggplot')
    
    # skewness and kurtosis
    print("Skewness (degree of distortion from the symmetrical normal curve):                " + str(df[feature].skew()))
    print("Kurtosis (it is the measure of the extreme values(outliers) in the distribution): " + str(df[feature].kurt()))

    ## Creating a customized chart. and giving in figsize and everything. 
    fig = plt.figure(constrained_layout=True, figsize=(12,8))
    ## creating a grid of 3 cols and 3 rows. 
    grid = gridspec.GridSpec(ncols=3, nrows=3, figure=fig)
    #gs = fig3.add_gridspec(3, 3)

    ## Customizing the histogram grid. 
    ax1 = fig.add_subplot(grid[0, :2])
    ## Set the title. 
    ax1.set_title('Histogram')
    ## plot the histogram. 
    sns.distplot(df.loc[:,feature], norm_hist=True, ax = ax1)

    # customizing the QQ_plot. 
    ax2 = fig.add_subplot(grid[1, :2])
    ## Set the title. 
    ax2.set_title('QQ_plot')
    ## Plotting the QQ_Plot. 
    stats.probplot(df.loc[:,feature], plot = ax2)

    ## Customizing the Box Plot. 
    ax3 = fig.add_subplot(grid[:, 2])
    ## Set title. 
    ax3.set_title('Box Plot')
    ## Plotting the box plot. 
    sns.boxplot(df.loc[:,feature], orient='v', ax = ax3 );
```

```python
# plot
plotting_3_chart(df, 'SalePrice')
```

One of the assumptions to apply linear models is the **linearality** between label and features. Another one is **Homoscedasticity** ( Constant Variance ): The assumption of Homoscedasticity is crucial to linear regression models. Homoscedasticity describes a situation in which the error term or variance or the "noise" or random disturbance in the relationship between the independent variables and the dependent variable is the same across all values of the independent variable. In other words, there is a constant variance present in the response variable as the predictor variable increases. If the "noise" is not the same across the values of an independent variable like the residual plot above, we call that Heteroscedasticity. As you can tell, it is the opposite of Homoscedasticity. Here is an excellent example of Homoscedasticity



```python
# FUNCTION TO PLOT BOX-PLOT o SCATTERPLOT
def plot_boxplot(n,m,df,figx,figy,y_val,box=1):
    """This function plot a multple subplot (defined by n & m) for a dataframe, also defining the size of the figure.
    The plot could be boxplot (box=1) for categorical data, and regplot for numerical data"""
    fig, axs = plt.subplots(n,m,figsize=(figx,figy))
    rect = fig.patch
    rect.set_facecolor('white')
    axs = axs.flatten()
    
    ll= df.columns.to_list()
    ll.remove(y_val)

    for i,(ax, feature) in enumerate (zip(axs, ll)):
        if box == 1:
            boxplot= sns.boxplot(data=df,x=feature, y=y_val,ax=ax)
        elif box==0:
            #boxplot= sns.scatterplot(data=df,x=feature, y='SalePrice',ax=ax,alpha=0.8)
            ## Putting a regression line. 
            boxplot = sns.regplot(data=df,x=feature, y=y_val, ax=ax,
                                  scatter_kws = {'color': 'gray', 'alpha':'0.5'}, 
                                  line_kws = {'color': 'blue'} )
        elif box==2:
            try:
                boxplot=sns.distplot(df[feature], ax=ax)
            except: 
                pass
            
        boxplot.set_xticklabels(boxplot.get_xticklabels(), rotation=45, horizontalalignment='right')
    plt.tight_layout()
```

```python
# FUNCTION TO CHECK FOR NANS AND THEIR RELATIVE %
def missing_percentage(df):
    """This function takes a DataFrame(df) as input and returns two columns, total missing values and total missing values percentage"""
    total = df.isnull().sum().sort_values(ascending = False)[df.isnull().sum().sort_values(ascending = False) != 0]
    percent = round(df.isnull().sum().sort_values(ascending = False)/len(df)*100,2)[round(df.isnull().sum().sort_values(ascending = False)/len(df)*100,2) != 0]
    return pd.concat([total, percent], axis=1, keys=['Total','Percent'])
```

```python
cor = df.corr()
#Correlation with output variable
cor_target = abs(cor["SalePrice_log10"])
#Selecting highly correlated features
relevant_features = cor_target[cor_target>0.0]
print ('Relevant Features used in the model:\n\n',relevant_features.sort_values(ascending=False) )
```

```python
list_relevant_features.remove('Id')
list_relevant_features.remove('SalePrice')
list_relevant_features.remove('SalePrice_log10')
```

```python
# USE ONLY RELEVANT FEATURES
df_ML = df[list_relevant_features]
#df_ML.drop(columns=['SalePrice','SalePrice_log10','Id'],inplace=True)

df_ML_test = df_test[list_relevant_features]
#df_ML_test.drop(columns=['Id'],inplace=True)

relevant_features = df_ML.columns.to_list()

# NORMALIZAMOS FEATURES
pt_ML = PowerTransformer()
pt_ML.fit(df_ML[relevant_features])

df_ML.loc[:,relevant_features] = pt_ML.transform(df_ML[relevant_features])
df_ML_test.loc[:,relevant_features] = pt_ML.transform(df_ML_test[relevant_features])

# feature 
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler().fit(df_ML)
#transformer = RobustScaler().fit(df_ML)
df_ML.loc[:,relevant_features] = sc.transform(df_ML)

df_ML_test.loc[:,relevant_features] = sc.transform(df_ML_test)
```

```python
from sklearn.ensemble import RandomForestRegressor

# SEGUNDO MODELO MULTILINEAL I ALGUNOS RESULTADOS
X_train, X_test, y_train, y_test = train_test_split(df_ML, y, test_size=0.33, random_state=42)

# Modelo
ML=LinearRegression(n_jobs=-1)
RF=RandomForestRegressor(n_estimators=300,n_jobs=-1)


#CV R2 
results_r2 = cross_val_score(ML, X_train, y_train, cv=6, scoring='r2')
results_rms = cross_val_score(ML, X_train, y_train, cv=6, scoring='neg_root_mean_squared_error')

print (' ########      CROSS VALIDATION LR     ############## ')
print('R^2 (coefficient of determination) regression score function for the cross validation:',results_r2.mean())
print('Root Mean Squared Error (log data) score function for the cross validation:',-results_rms.mean())

# FIT AND PREDICT
ML.fit(X_train,y_train)
pred_lineal= ML.predict(X_test)
                  
# VALIDATION USING RMSE OF LOG SALES
print (' \n ########      TEST VALIDATION      ############## ')
print('Root Mean Squared Error (log data): ',rmsle_log(y_test, pred_lineal))
```
